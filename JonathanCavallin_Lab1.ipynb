```r
#7.1)
library("geosphere")

df_skimmed_one_hour <- filter(df_skimmed,tripduration<=3600) 

df_skimmed_one_hour <- mutate(df_skimmed_one_hour,velocity=distHaversine(cbind(`start station latitude`,`start station longitude`),cbind(`end station latitude`,`end station longitude`))/tripduration)
# distHaversine(p1,p2) with p1=(latitude,longitude) gives the distance between the points in meters

df_skimmed_one_hour$velocity<- df_skimmed_one_hour$velocity*3.6  # velocity in km/h 
df_skimmed_one_hour
```

```r
#7.2)
# interpreting in this way

options(repr.plot.width = 15, repr.plot.height = 8)

df_skimmed_one_hour<- mutate(df_skimmed_one_hour,distance= df_skimmed_one_hour$velocity*df_skimmed_one_hour$tripduration/3600) # add distance column in km

df_skimmed_one_hour_0_5 <- df_skimmed_one_hour[df_skimmed_one_hour$distance<=0.5,]
df_skimmed_one_hour_1_0 <- df_skimmed_one_hour[df_skimmed_one_hour$distance>0.5 & df_skimmed_one_hour$distance<=1.0,]
df_skimmed_one_hour_2_0 <- df_skimmed_one_hour[df_skimmed_one_hour$distance>1.0 & df_skimmed_one_hour$distance<=2.0,]
df_skimmed_one_hour_3_0 <- df_skimmed_one_hour[df_skimmed_one_hour$distance>2.0 & df_skimmed_one_hour$distance<=3.0,]
df_skimmed_one_hour_4_0 <- df_skimmed_one_hour[df_skimmed_one_hour$distance>3.0,]

ggplot(df_skimmed_one_hour_0_5, aes(x = distance, y = velocity)) +
  geom_point(color='blue')+
 labs(x = " Route length (km)", y = "Average velocity (km/h)", title = "Average velocity vs distance covered (d<0.5km)")

ggplot(df_skimmed_one_hour_1_0, aes(x = distance, y = velocity)) +
  geom_point(color='blue')+
 labs(x = " Route length (km)", y = "Average velocity (km/h)", title = "Average velocity vs distance covered (0.5km<d<1km)")

ggplot(df_skimmed_one_hour_2_0, aes(x = distance, y = velocity)) +
  geom_point(color='blue')+
 labs(x = " Route length (km)", y = "Average velocity (km/h)", title = "Average velocity vs distance covered (1km<d<2km)")

ggplot(df_skimmed_one_hour_3_0, aes(x = distance, y = velocity)) +
  geom_point(color='blue')+
 labs(x = " Route length (km)", y = "Average velocity (km/h)", title = "Average velocity vs distance covered (2km<d<3km)")

ggplot(df_skimmed_one_hour_4_0, aes(x = distance, y = velocity)) +
  geom_point(color='blue')+
 labs(x = " Route length (km)", y = "Average velocity (km/h)", title = "Average velocity vs distance covered (d>3km)")
```

```r
# other interpretation
avg_vel_0_5 <- df_skimmed_one_hour_0_5 |> summarise(m_velocity= mean(velocity),std=sd(velocity))
avg_vel_1_0 <- df_skimmed_one_hour_1_0 |> summarise(m_velocity= mean(velocity),std=sd(velocity))
avg_vel_2_0 <- df_skimmed_one_hour_2_0 |> summarise(m_velocity= mean(velocity),std=sd(velocity))
avg_vel_3_0 <- df_skimmed_one_hour_3_0 |> summarise(m_velocity= mean(velocity),std=sd(velocity))
avg_vel_4_0 <- df_skimmed_one_hour_4_0 |> summarise(m_velocity= mean(velocity),std=sd(velocity))

plot_data <- data.frame(
  d = c(0.25, 0.75, 1.5, 2.5, 3.5),  # The distance bins
  velocity = c(avg_vel_0_5$m_velocity, avg_vel_1_0$m_velocity, avg_vel_2_0$m_velocity, avg_vel_3_0$m_velocity, avg_vel_4_0$m_velocity),
  std = c(avg_vel_0_5$std, avg_vel_1_0$std, avg_vel_2_0$std, avg_vel_3_0$std, avg_vel_4_0$std),
  group = rep(c("0-0.5", "0.5-1.0", "1.0-2.0", "2.0-3.0", "3.0+"), each = 5)  
    )# Grouping by the bins


ggplot2(plot_data, aes(x = d, y = velocity,color='red')) +
  geom_point() +  
  geom_line() +   
  geom_errorbar(aes(ymin = velocity - std, ymax = velocity + std), width = 0.1) +  
  labs(x = "Distance (km)", 
       y = "Average Velocity (km/h)", 
       title = "Average Velocity vs route distance with std dev") +
  theme(legend.position = "top")
```

```r
#1.2)
# for each event and for each plane find maximum and second maximum pixel values

df_max <- df |>  mutate(
max_pixel_1 = apply(df[, 1:64], 1, function(x) max(x, na.rm = TRUE)),  # 1 tells apply to operate row-wise
second_max_pixel_1 = apply(df[, 1:64], 1, function(x) {
      sorted_x <- sort(x, decreasing = TRUE, na.last = NA)  
      if (length(sorted_x) >= 2) {
        return(sorted_x[2])  # return the second maximum
      } else {
        return(NA)  # return NA if there’s no second max
      }
    }),
max_pixel_2 = apply(df[,65:128], 1, function(x) max(x, na.rm = TRUE)),  
 second_max_pixel_2 = apply(df[, 65:128], 1, function(x) {
      sorted_x <- sort(x, decreasing = TRUE, na.last = NA)  
      if (length(sorted_x) >= 2) {
        return(sorted_x[2])  
      } else {
        return(NA)  
      }
    })  )
   
df_max
```

```r
#7) spatial distribution heatmaps

all_max_idx <- df_max_positive_1$id_max_1  
all_max_idx_2 <- df_max_positive_2$id_max_2  

max_counts <- as.data.frame(table(all_max_idx))   # table() counts how many times each unique value appears in a vector
max_counts_2 <- as.data.frame(table(all_max_idx_2))


colnames(max_counts) <- c("idx", "count")
colnames(max_counts_2) <- c("idx", "count")

max_counts$idx <- as.numeric(as.character(max_counts$idx))
max_counts_2$idx <- as.numeric(as.character(max_counts_2$idx))

# compute row and column numbers
max_counts <- max_counts |>
  mutate(
    row = idx %% 8,
    col = idx %/% 8
  )

max_counts_2 <- max_counts_2 |>
  mutate(
    row = idx %% 8,
    col = idx %/% 8
  )

#convert to 1-8 
max_counts <- max_counts |>
  mutate(row = row + 1, col = col + 1)
max_counts_2 <- max_counts_2 |>
  mutate(row = row + 1, col = col + 1)


heatmap_plot <- ggplot(max_counts, aes(x = col, y = row, fill = count)) +
  geom_tile(color = "black") +
  scale_fill_gradient(low = "lightblue", high = "red") +
  scale_y_reverse() +  
  coord_fixed() +
  labs(title = "Spatial distribution of max pixel index: upstream matrix",
       x = "Column", y = "Row", fill = "Frequency") 

heatmap_plot_2 <- ggplot(max_counts_2, aes(x = col, y = row, fill = count)) +
  geom_tile(color = "black") +
  scale_fill_gradient(low = "lightblue", high = "red") +
  scale_y_reverse() + 
  coord_fixed() +
  labs(title = "Spatial distribution of max pixel index: downstream matrix",
       x = "Column", y = "Row", fill = "Frequency") 


grid.arrange(heatmap_plot,heatmap_plot_2,nrow=1,ncol=2)
```
