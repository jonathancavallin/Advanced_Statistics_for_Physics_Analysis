```r
#6.1)
df_skimmed$starttime <- as.POSIXct(df_skimmed$starttime) 
df_skimmed <- mutate(df_skimmed,day=format(df_skimmed$starttime, "%A"))  # %A gives the day of the week
df_skimmed
```

```r
number_rides <- df_skimmed |> group_by(day) |> summarise(rides = length(tripduration)) # summarise creates a new dataframe with a column named rides and idx day on which we grouped 
number_rides

number_rides$day <- factor(number_rides$day, levels = c("Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"))

options(repr.plot.width = 15, repr.plot.height = 8)

ggplot(number_rides, aes(x = day, y = rides)) +
  geom_bar(stat = "identity", fill = "orange",color="black") +
  labs(x = "Day", y = "Number of rides", title = "Number of rides per day")
```

```r
# or equivalently
ggplot(df_skimmed, aes(x = day)) +
  geom_bar(stat = "count", fill = "orange",color="black") +
  labs(x = "Day", y = "Number of rides", title = "Number of rides per day") # stat= count counts the number of elemnt of each category in column day
```

```r
#6.2)
df_skimmed <- mutate(df_skimmed,hour=format(df_skimmed$starttime, "%H")) # %H gives the hour of the day of UNIX format object 
df_skimmed
```

```r
df_skimmed_weekends <- filter(df_skimmed, day=='Saturday' | day=='Sunday' ) # | is OR
df_skimmed_weekdays <- filter(df_skimmed, day!='Saturday' & day!='Sunday' ) 

plot_weekdays <- ggplot(df_skimmed_weekdays, aes(x=hour))+
geom_bar(stat='count',fill='green',color='black')+
  labs(x = "Hour", y = "Number of rides", title = "Number of rides per hour: weekdays")

plot_weekends <-ggplot(df_skimmed_weekends, aes(x=hour))+
geom_bar(stat='count',fill='red',color='black')+
  labs(x = "Hour", y = "Number of rides", title = "Number of rides per hour: weekends")

options(repr.plot.width = 20, repr.plot.height = 8)
library(gridExtra)
grid.arrange(plot_weekdays,plot_weekends,nrow=1,ncol=2)
```

We can clealy see the two peaks in the weekdays, corresponding to the commute times for going to and returning from work, which are absent during the weekends.

```r
#6.3)
df_skimmed_weekdays_cust <- filter(df_skimmed_weekdays, usertype == 'Customer')
df_skimmed_weekdays_sub <- filter(df_skimmed_weekdays, usertype == 'Subscriber')

plot_weekdays_cust <- ggplot(df_skimmed_weekdays_cust, aes(x=hour))+
geom_bar(stat='count',fill='darkgreen',color='black')+
  labs(x = "Hour", y = "Number of rides", title = "Number of rides per hour on weekdays: Customers")

plot_weekdays_sub <-ggplot(df_skimmed_weekdays_sub, aes(x=hour))+
geom_bar(stat='count',fill='lightgreen',color='black')+
  labs(x = "Number", y = "Number of rides", title = "Number of rides per hour on weekdays: Subscribers")
grid.arrange(plot_weekdays_sub,plot_weekdays_cust,nrow=1,ncol=2)
```

As expected subscribers are more likely to be workers, with a hourly distribution similar to the global on weekdays, while occasional customers have a distribution which is more similar to the one on weekends. 

```r
#7.1)
library("geosphere")

df_skimmed_one_hour <- filter(df_skimmed,tripduration<=3600) 

df_skimmed_one_hour <- mutate(df_skimmed_one_hour,velocity=distHaversine(cbind(`start station latitude`,`start station longitude`),cbind(`end station latitude`,`end station longitude`))/tripduration)
# distHaversine(p1,p2) with p1=(latitude,longitude) gives the distance between the points in meters

df_skimmed_one_hour$velocity<- df_skimmed_one_hour$velocity*3.6  # velocity in km/h 
df_skimmed_one_hour
```

```r
#7.2)
# interpreting in this way

df_skimmed_one_hour<- mutate(df_skimmed_one_hour,distance= df_skimmed_one_hour$velocity*df_skimmed_one_hour$tripduration/3600) # add distance column in km

df_skimmed_one_hour_0_5 <- df_skimmed_one_hour[df_skimmed_one_hour$distance<=0.5,]
df_skimmed_one_hour_1_0 <- df_skimmed_one_hour[df_skimmed_one_hour$distance>0.5 & df_skimmed_one_hour$distance<=1.0,]
df_skimmed_one_hour_2_0 <- df_skimmed_one_hour[df_skimmed_one_hour$distance>1.0 & df_skimmed_one_hour$distance<=2.0,]
df_skimmed_one_hour_3_0 <- df_skimmed_one_hour[df_skimmed_one_hour$distance>2.0 & df_skimmed_one_hour$distance<=3.0,]
df_skimmed_one_hour_4_0 <- df_skimmed_one_hour[df_skimmed_one_hour$distance>3.0,]

ggplot(df_skimmed_one_hour_0_5, aes(x = distance, y = velocity)) +
  geom_point(color='blue')+
 labs(x = " Route length (km)", y = "Average velocity (km/h)", title = "Average velocity vs distance (d<0.5km)")

ggplot(df_skimmed_one_hour_1_0, aes(x = distance, y = velocity)) +
  geom_point(color='blue')+
 labs(x = " Route length (km)", y = "Average velocity (km/h)", title = "Average velocity vs distance (0.5km<d<1km)")

ggplot(df_skimmed_one_hour_2_0, aes(x = distance, y = velocity)) +
  geom_point(color='blue')+
 labs(x = " Route length (km)", y = "Average velocity (km/h)", title = "Average velocity vs distance (1km<d<2km)")

ggplot(df_skimmed_one_hour_3_0, aes(x = distance, y = velocity)) +
  geom_point(color='blue')+
 labs(x = " Route length (km)", y = "Average velocity (km/h)", title = "Average velocity vs distance (2km<d<3km)")

ggplot(df_skimmed_one_hour_4_0, aes(x = distance, y = velocity)) +
  geom_point(color='blue')+
 labs(x = " Route length (km)", y = "Average velocity (km/h)", title = "Average velocity vs distance (d>3km)")
```

```r
# other interpretation
avg_vel_0_5 <- df_skimmed_one_hour_0_5 |> summarise(m_velocity= mean(velocity),std=sd(velocity))
avg_vel_1_0 <- df_skimmed_one_hour_1_0 |> summarise(m_velocity= mean(velocity),std=sd(velocity))
avg_vel_2_0 <- df_skimmed_one_hour_2_0 |> summarise(m_velocity= mean(velocity),std=sd(velocity))
avg_vel_3_0 <- df_skimmed_one_hour_3_0 |> summarise(m_velocity= mean(velocity),std=sd(velocity))
avg_vel_4_0 <- df_skimmed_one_hour_4_0 |> summarise(m_velocity= mean(velocity),std=sd(velocity))

plot_data <- data.frame(
  d = c(0.25, 0.75, 1.5, 2.5, 3.5),  # The distance bins
  velocity = c(avg_vel_0_5$m_velocity, avg_vel_1_0$m_velocity, avg_vel_2_0$m_velocity, avg_vel_3_0$m_velocity, avg_vel_4_0$m_velocity),
  std = c(avg_vel_0_5$std, avg_vel_1_0$std, avg_vel_2_0$std, avg_vel_3_0$std, avg_vel_4_0$std),
  group = rep(c("0-0.5", "0.5-1.0", "1.0-2.0", "2.0-3.0", "3.0+"), each = 5)  s
    )# Grouping by the bins

ggplot(plot_data, aes(x = d, y = velocity, color = group)) +
  geom_point() +  
  geom_line() +   
  geom_errorbar(aes(ymin = velocity - std, ymax = velocity + std), width = 0.1) +  
  labs(x = "Distance (km)", 
       y = "Average Velocity (km/h)", 
       title = "Average Velocity vs route distance with std dev") +
  theme_minimal() +
  theme(legend.position = "top")
```
